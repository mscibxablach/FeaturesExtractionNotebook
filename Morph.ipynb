## Imports

```python
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

## Show image

```python
def show_image(title, image):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

## Load image

```python
def read_image(filename):
    return cv2.imread(filename)

filename = 'Photos/43.jpg'
image = read_image(filename)

show_image('image', image)
```

## Threshold

```python
def threshold(image):
    image = image.copy()
    return cv2.threshold(image, 70, 255, cv2.THRESH_BINARY)

ret, thresh = threshold(image)
show_image('image', image)
show_image('threshold', thresh)
    
```

## Erosion

```python
def erosion(image):
    image = image.copy()
    kernel = np.ones((5,5), np.uint8)
    return cv2.erode(image, kernel, iterations=1)

_, thresh = threshold(image)
thresh_erosion_result = erosion(thresh)
erosion_result = erosion(image)
show_image('image', image)
show_image('erosion', erosion_result)
show_image('thresh erosion', thresh_erosion_result)
```

## Dilation

```python
def dilate(image):
    image = image.copy()
    kernel = np.ones((3,3), np.uint8)
    return cv2.dilate(image, kernel, iterations=1)

_, thresh = threshold(image)
dilate_thresh = dilate(thresh)
dilate_result = dilate(image)

show_image('image', image)
show_image('dilate result', dilate_result)
show_image('thresh dilate', dilate_thresh)
```

## Opening

```python
def opening(image):
    image = image.copy()
    kernel = np.ones((5,5), np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

_, thresh = threshold(image)
opening_thresh = opening(thresh)
opening_result = opening(image)

show_image('image', image)
show_image('opening thresh', opening_thresh)
show_image('opening result', opening_result)
```

## Closing

```python
def closing(image):
    image = image.copy()
    kernel = np.ones((5,5), np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

_, thresh = threshold(image)
closing_thresh = closing(thresh)
closing_result = closing(image)

show_image('image', image)
show_image('closing result', closing_result)
show_image('closing thresh', closing_thresh)
```

## Thresh-opening-closing

```python
_, thresh = threshold(image)
opening = opening(thresh)
closing = closing(opening)
show_image('image', image)
show_image('result', closing)
```

## Opening-closing

```python
opening1 = opening(image)
closing1 = closing(opening1)

show_image('image', image)
show_image('result', closing1)
```
